```python
plt.xkcd()

plt.title('Bug hunt')

x = arange(5)
bugs = [1200, 1190, 800, 350, 200]

def kilos(x, pos):
    'The two args are the value and tick position'
    return '%1.1fk' % (x*1e-3)

formatter = FuncFormatter(kilos)

ax = subplot(111)
ax.yaxis.set_major_formatter(formatter)
ax.bar(x, bugs, label = '#bugs')

plt.xlabel('Release')
xticks( x,  ('1.3', '1.3.1', '2.0', '2.1', '2.1.1') )

# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axhline.html#matplotlib-axes-axes-axhline
plt.axhline(y=250, color='red', linestyle='--', label='threshold')

plt.legend()

plt.savefig('/home/jovyan/work/bars/bugs-over-time.svg')
```
