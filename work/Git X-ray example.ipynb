```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])

    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 240])
    ax.set_xticklabels(['4 years, 11 months ago', '7 weeks ago'])
    ax.set_xlim([-25, 260])

    plt.text(0, 1, 'migration from\n Pebble to Pelican', withdash=True,
        dashdirection=1,
               dashpad=10, 
               dashlength=25,
             dashpush=10,
               dashrotation=45)

    plt.text(240, 1, 'last post', withdash=True,
    dashdirection=1,
           dashpad=10, 
           dashlength=25,
         dashpush=10,
           dashrotation=155)

    plt.text(250, 1, 'last change', withdash=True,
        dashdirection=0,
               dashpad=10, 
               dashlength=25,
             dashpush=10,
               dashrotation=65)

data = np.array([
    [0, 1],
    [240, 1],
    [250, 1],
])
x, y = data.T
plt.scatter(x,y)

plt.title('Blog Events')
plt.savefig('/home/jovyan/work/xray/step4-labels.svg')
```
