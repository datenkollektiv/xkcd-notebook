```python
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

import pandas as pd

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
print(df.shape)
```

```python
X_sepalLengh = df['sepal length (cm)'].values.reshape(-1,1)
print (X_sepalLengh.shape)
```

```python
Y_sepalWidth = df['sepal width (cm)'].values
print (Y_sepalWidth.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_sepalLengh, Y_sepalWidth)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_sepalLengh, Y_sepalWidth)
prediction = regression.predict(X_sepalLengh)

plt.scatter(X_sepalLengh, Y_sepalWidth)
plt.plot(X_sepalLengh, prediction, color='orange')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

The result is `y = ax + b` with a and b being the parameters/coefficients of the model or

* `a` the slope and
* `b` the intercept.
