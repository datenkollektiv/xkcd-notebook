```python
plt.xkcd()

columns = ('module A', 'module B', 'module C', 'module D')
rows = ['clean-up', 'jdbc', 'services', 'clean-up', 'e4']

data = [[ 100,  90,  90,  80],
        [  90,  60,  30,   0],
        [  80,  50,  20,   0],
        [  70,  40,  10,   0],
        [  60,  30,   0,   0]]

# Get some pastel shades for the colors
n_rows = len(data)

index = np.arange(len(columns)) + 0.3
bar_width = 0.7

# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns))

# Plot bars and create text labels for the table
cell_text = []
colors = []

for row in range(n_rows):
    # TODO - replace data[row] with constants to create the blocks
    # TODO - calculate colors based on data[row]
#     colors = np.zeros(len(data[row]))
    colors = []
    for column in range(len(data[row])):
        if (data[row][column]) < 20:
            colors.append('red')
        elif (data[row][column]) <40:
            colors.append('orangered')
        elif (data[row][column]) <60:
            colors.append('orange')
        elif (data[row][column]) <80:
            colors.append('gold')
        elif (data[row][column]) <100:
            colors.append('yellowgreen')
        else:
            colors.append('green')

    plt.bar(index, np.ones_like(data[row]), bar_width, bottom=y_offset, color=colors)
    y_offset = y_offset + 1
    cell_text.append(['%1.1f %s' % (x, '%') for x in data[row]])

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      colLabels=columns,
                      loc='bottom')
the_table.auto_set_font_size(False)
the_table.set_fontsize(8)

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.2, bottom=0.25)

plt.xticks([])
plt.yticks(np.arange(start=0.5, stop=5.5, step=1), ('clean-up', 'jdbc', 'services', 'clean-up', 'e4'))

plt.axhline(y=3, color='grey', linestyle='--', label='REST')
plt.legend()

plt.annotate(r"$\}$",fontsize=32,
            xy=(0.90, 0.70), xycoords='figure fraction'
            )
plt.annotate(r"$UI$",fontsize=18,
            xy=(0.95, 0.71), xycoords='figure fraction'
            )

plt.title('Management Overview')
plt.savefig('/home/jovyan/work/bars/management-overview.svg')
```
