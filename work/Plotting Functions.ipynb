```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='f(x) = x')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title("Identity function")
plt.legend()

plt.savefig('/home/jovyan/work/axis/identity-function-boxed.svg')
```

```python
x = np.linspace(-2, 2, 100)

plt.plot(x, x*x*x, label='f(x) = x^3')

plt.title("Power function")
plt.legend()

plt.axis('off')
plt.axis('equal')
plt.axis([-3, 3, -3, 3])

plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')

plt.savefig('/home/jovyan/work/axis/power-function-plain.svg')
```

```python
from mpl_toolkits.axes_grid.axislines import SubplotZero
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(1)
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

x = np.linspace(-0.5, 1., 100)
ax.plot(x, np.sin(x*np.pi))

plt.title("Sinus function")
plt.legend(['f(x) = sin(x)'])

plt.savefig('/home/jovyan/work/axis/sin-function-axislines.svg')
```
